```python
"""This program collects info from two links I found that contained HTML-tables of
the Paldea Pokedex with the following info: 
Pokedex number, pokemon name, type and location/evolution.

The first link is the one I preffered, but it lacked some info of location,
due to it using images instead of text which I wanted. 
Therefor I found another link that had this info, which I will merge into 
one final list and save to txt-file.

The txt-file will be used in iOS(Swift) app to gather data from."""
```

```python
import pandas as pd
import numpy as np
```

```python
# Info link:
# https://www.eurogamer.net/pokemon-scarlet-violet-pokedex-all-locations-paldea-listed-9015

# Additional info - > will be replacing "nan" from first link:
# https://www.ign.com/wikis/pokemon-scarlet-violet/Pokedex_-_All_Pokemon_in_Scarlet_and_Violet
```

```python
data = pd.read_html('https://www.eurogamer.net/pokemon-scarlet-violet-pokedex-all-locations-paldea-listed-9015')[0]

#Deleting column with photos - we are only looking for text info at this point
del data['Unnamed: 4'] 

data
```

```python
additionalData = pd.read_html('https://www.ign.com/wikis/pokemon-scarlet-violet/Pokedex_-_All_Pokemon_in_Scarlet_and_Violet')[0]

#Additional info had some reapeting pokemon -> we delete these as not needed
additionalData = additionalData.drop([223]).reset_index(drop = True)
additionalData = additionalData.drop([223]).reset_index(drop = True)

additionalData
```

```python
dataLst = pd.DataFrame.to_numpy(data)
addLst = pd.DataFrame.to_numpy(additionalData)
```

```python
#Replacing missing info(previously nan), with info from additional link.

for i in range(len(dataLst)):
    if type(dataLst[i][-1]) == type(np.nan):
        new = np.char.replace(addLst[i][-1], "etc.", "++")
        dataLst[i][-1] = str(new)
```

```python
#Saving end result to txt file.

np.savetxt("paldeaData.txt", dataLst, fmt = "%s;")
```
